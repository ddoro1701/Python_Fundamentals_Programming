```python
def find_max_min(arr):
    if not arr:
        return None

    max_val = arr[0]
    min_val = arr[0]

    for num in arr:
        if num > max_val:
            max_val = num
        elif num < min_val:
            min_val = num
    return (max_val, min_val)
```

```python
# Test the function
input_array = [3, 1, 7, 9, 4, 2, 100]
result = find_max_min(input_array)
print("Maximum and Minimum values: ", result)
```
