```python
%pip install networkx
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge(1, 2)
nx.draw_networkx(G)
plt.show()
```

```python
G.add_nodes_from([3, 4])
nx.draw_networkx(G)
plt.show()
```

```python
G.add_edge(3, 4)
G.add_edges_from([(2,3), (4, 1)])
nx.draw_networkx(G)
plt.show()
```
