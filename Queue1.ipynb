```python
#initializing a queue
queue = []
```

```python
queue.append('a')
queue.append('b')
queue.append('c')
```

```python
print("Initial Queue")
```

```python
print(queue)
```

```python
# Removing elements from the queue
print("\nElements dequeued from a queue")
print(queue.pop(1))
print(queue.pop(0))
print(queue.pop(0))
```

```python
print("\nQueue after removing elements")
print(queue)
```

```python
from queue import Queue
#initializing a queue
q = Queue(maxsize = 3)
```

```python
# qsize give the maxsize of the Queue
print(q.qsize())
```

```python
# Adding of element to queue
q.put('a')
q.put('b')
q.put('c')
```

```python
#Return boolean for full Queue
print("\nElements dequeued from the queue")
print(q.get())
print(q.get())
print(q.get())
```

```python
#Return Boolean for Empty
#Queue
print("\nEmpty: ", q.empty())
```

```python
q.put(1)
print("\nEmpty: ", q.empty())
print("Full: ", q.full())

#This would result into Infinite
#Loop as the Queue is empty
# print(q.get())
```
