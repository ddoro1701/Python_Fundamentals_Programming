```python
%%time
# Reverse a Python list with a List Comprehension
original_list = [1,2,3,4,5,6,7,8,9]
```

```python
start = len(original_list) - 1
```

```python
print (original_list)
```

```python
reversed_list = [original_list[i] for i in range (start, -1, -1)]
```

```python
print (reversed_list)
```

```python
%%time
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)
fib(32)
```
